# ***BREAST CANCER PREDICTION***

***ABOUT THE DATASET***





**PURPOSE:** To classify breast cancer tumors as malignant or benign based on features extracted from mammograms.

***Columns in the Dataset***

**1] Id :** Unique identifier for each patient (not useful for prediction).

**2] Diagnosis :** The target variable indicating the diagnosis:

M for malignant

B for benign

**3] Radius mean :** Mean radius of the tumors.

**4] Texture mean :** Mean texture of the tumors.

**5] Perimeter mean :** Mean perimeter of the tumors.

**6] Area mean :** Mean area of the tumors.

**7] Smoothness mean :** Mean smoothness of the tumors.

**8] Compactness mean :** Mean compactness of the tumors.

**9] Concavity mean :** Mean concavity of the tumors.

**10] Concave points mean :** Mean number of concave points of the tumors.

**11] Symmetry mean :** Mean symmetry of the tumors.

**12] Fractal dimension mean :** Mean fractal dimension of the tumors.

**13] Radius se:** Standard error of the radius.

**14] Texture se :** Standard error of the texture.

**15] Perimeter se :** Standard error of the perimeter.

**16] Area se :** Standard error of the area.

**17] Smoothness se:** Standard error of the smoothness.

**18] Compactness se :** Standard error of the compactness.

**19] Concavity se :** Standard error of the concavity.

**20] Concave points se :** Standard error of the concave points.

**21] Symmetry se :** Standard error of the symmetry.

**22] Fractal dimension se :** Standard error of the fractal dimension.

**23] Radius worst :** Worst (largest) radius of the tumors.

**24] Exture worst :** Worst texture of the tumors.


**25] Perimeter worst :** Worst perimeter of the tumors

**26] Area worst :** Worst area of the tumors.

**27] Smoothness worst** : Worst smoothness of the tumors.

**28] Compactness worst :** Worst compactness of the tumors.

**29] Concavity worst :** Worst concavity of the tumors.

**30] Concave points worst :** Worst concave points of the tumors.

**31] Symmetry worst :** Worst symmetry of the tumors.

**32] Fractal dimension worst :** Worst fractal dimension of the tumors.

**33] Unnamed: 32 :** This is often an extra column that doesn't contain any useful data.



***Data Characteristics***

Number of Instances : 569

Number of Features : 30 (excluding the 'id' and 'Unnamed: 32' columns)

Target Variable : diagnosis (malignant or benign)

Feature Types : Numeric (all features are continuous)

***COLLECTING INPUT AND OUTPUT DATA***

```python
x=df.drop(['diagnosis_full'],axis=1).values
x
```

```python
y=df['diagnosis_full'].values
y
```

***SPLITTING DATA INTO TRAINING AND TESTING***

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.30,random_state=42)
```

***NORMALIZATION USING STANDARD SCALER***

```python
scaler=StandardScaler()
scaler.fit(x_train)
x_train= scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

***MODEL CREATION USING ALGORITHMS***

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

***a] KNeighborsClassifier***

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
knn_predict =knn.predict(x_test)
knn_acc = accuracy_score(knn_predict,y_test)
print("Test accuracy of RandomForestClassifier model is:{:2f}%".format(knn_acc*100))
```

```python
print(confusion_matrix(y_test,knn_predict))
print("*"*100)
print(classification_report(y_test,knn_predict))
```

***b] Random Forest Classifier***

```python
Random = RandomForestClassifier(n_estimators=100, random_state=42)
Random.fit(x_train,y_train)
Random_predict =Random.predict(x_test)
Random_acc = accuracy_score(Random_predict,y_test)
print("Test accuracy of RandomForestClassifier model is:{:2f}%".format(Random_acc*100))
```

```python
print(confusion_matrix(y_test,Random_predict))
print("*"*100)
print(classification_report(y_test,Random_predict))
```

***c] Decision Tree Classifier***

```python
tree = DecisionTreeClassifier(random_state=42)
tree.fit(x_train,y_train)
tree_predict =tree.predict(x_test)
tree_acc = accuracy_score(tree_predict,y_test)
print("Test accuracy of DecisionTreeClassifier model is:{:2f}%".format(tree_acc*100))
```

```python
print(confusion_matrix(y_test,tree_predict))
print("*"*100)
print(classification_report(y_test,tree_predict))
```

***d] Logistic Regression***

```python
lg = LogisticRegression()
lg.fit(x_train,y_train)
lg_predict =lg.predict(x_test)
lg_acc = accuracy_score(lg_predict,y_test)
print("Test accuracy of LogisticRegression model is :{:2f}%".format(lg_acc*100))
```

```python
print(confusion_matrix(y_test,lg_predict))
print("*"*100)
print(classification_report(y_test,lg_predict))
```
